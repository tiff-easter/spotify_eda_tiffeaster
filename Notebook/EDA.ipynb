#identify duplicated information
    #True- IS duplicated
    #False- is NOT duplicated


```python
df.duplicated()
```

#using value of 'keep' with 'duplicated' function
    #'first'(default) returns all duplicates as 'True' except for the first occurrence, which is marked 'False'
    #'last' returns all duplicates as 'True" except for the last occurrence, which is marked 'False'
    #'false' returns 'True' for all occurrences of rows that are duplicates

```python
df.duplicated(keep='last')
```

#add 'sum' function to see the number of duplicated values

```python
df.duplicated(keep=False).sum()
```

#use the 'loc' operator to pull the rows where the data is duplicated and look at each of those rows

```python
df.loc[df.duplicated(keep=False)]
```

##add 'sort values' for "Track_ID"
    #makes rows easier to compare
#finding duplicates will help determine which values need to be dropped or kept

```python
df.loc[df.duplicated(keep=False)].sort_values('track_id')
```

#multi variant analysis compares two of more variables to determine how they relate to each other using 'corr' function
    #first reduce columns to quanitative variables, since 'corr' only works on numeric data

```python
df.columns
df_corr = df[['popularity',
        'duration_ms', 'danceability', 'energy', 'key', 'loudness', 
        'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
        'valence', 'tempo', 'time_signature']].corr()
df_corr
```

#create heatmap

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df_corr)
```

#annotate heatmap and change color values

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df_corr, annot=True, cmap= 'mako')
```

#create pair plot

```python
sns.pairplot(df[['acousticness', 'energy']], height=5, aspect=1.5)
```

```python
df_sample = df[['acousticness', 'energy']].sample(n=5000, random_state=1)
sns.pairplot(df_sample, height=5, aspect=1.5, diag_kind='kde')
```
